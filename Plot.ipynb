```python
import matplotlib.pyplot as plt
import pandas as pd
import os
```

```python
base_path = os.getcwd()
months = list(range(1,4*12 + 10)) #4 anos + 10 meses
df_input = pd.read_csv(base_path+"/data/df_model_prediction_REPLICATION_YEAR.csv")
```

```python
df_bairro = df_input[df_input['nome_bairro'] == 'Bangu']
dengue_diagnosis = list(df_bairro["dengue_diagnosis"])
dengue_diagnosis_previsto = list(df_bairro["dengue_diagnosis_previsto"])
plt.plot(months,dengue_diagnosis)
plt.plot(months,dengue_diagnosis_previsto)
```
